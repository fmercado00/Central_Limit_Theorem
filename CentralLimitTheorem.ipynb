```python
# importing the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statistics import mean
```

```python
ds = pd.read_csv('dataset.csv')
sns.kdeplot(ds['Income'])
plt.show()
```

```python
OneThousandSamplesWith100Obs=[]
for i in range(0, 1001):
    subset = ds.sample(n=100)
    OneThousandSamplesWith100Obs.append(mean(subset['Income']))
```

```python
sns.kdeplot(OneThousandSamplesWith100Obs)
plt.show()
```
